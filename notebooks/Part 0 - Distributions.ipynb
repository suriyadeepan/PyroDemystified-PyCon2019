```python
import pyro.distributions as pdist
import torch
import seaborn as sns
import matplotlib.pyplot as plt
```

## Normal

$$x \sim \mathcal{N}(\mu, \sigma)$$

```python
normal = pdist.Normal(3, 1)
print(normal)
print('support', normal.support)
print('sample', normal.sample())
print('score (1)', normal.log_prob(1))
print('score (3)', normal.log_prob(3))
print('score (50)', normal.log_prob(50))
```

```python
sns.distplot(normal.sample([10000, ]))
```

## Caucy

```python
cauchy = pdist.Cauchy(10, 1)
print(cauchy)
print('support', cauchy.support)
print('sample', cauchy.sample())
print('score (1)', cauchy.log_prob(0.5))
print('score (3)', cauchy.log_prob(3))
print('score (50)', cauchy.log_prob(50))
```

```python
sns.distplot(cauchy.sample([100, ]))
```

## Half-Normal

```python
dist = pdist.HalfNormal(10)
print(dist)
print('support', dist.support)
print('sample', dist.sample())
print('score (1)', dist.log_prob(torch.tensor(0)))
print('score (3)', dist.log_prob(torch.tensor(10)))
print('score (50)', dist.log_prob(torch.tensor(500)))
```

```python
sns.distplot(dist.sample([10000, ]))
```

## Uniform

## Beta

```python
dist = pdist.Beta(12, 2)
print(dist)
print('support', dist.support)
print('sample', dist.sample())
print('score (1)', dist.log_prob(torch.tensor(0.)))
print('score (1)', dist.log_prob(torch.tensor(0.5)))
print('score (3)', dist.log_prob(torch.tensor(10.)))
print('score (50)', dist.log_prob(torch.tensor(500.)))
```

# Discrete

## Bernoulli

```python
dist = pdist.Bernoulli(0.8)
print(dist)
print('support', dist.enumerate_support())
print('sample', dist.sample())
```

```python
plt.hist(dist.sample([1000, ]))
```

## Binomial

```python
dist = pdist.Binomial(10, 0.5)
print(dist)
print('support', dist.enumerate_support())
print('sample', dist.sample())
```

```python
plt.hist(dist.sample([1000, ]))
```
